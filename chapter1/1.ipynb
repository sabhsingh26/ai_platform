```python
from typing import Union, Any
```

```python
def function1(a) -> any:
    return a * 2


def function2(b: int) -> int:
    return b + 5


def function3(c: Union[int, str]) -> Any:
    if isinstance(c, int):
        return c ** 2
    elif isinstance(c, str):
        return c.upper()
    else:
        return None


def function4(d: int | float) -> int | float:
    return d / 2.0

def named_function(param1: int, param2: str) -> str:
    return f"Parameter 1: {param1}, Parameter 2: {param2}"
```

```python
result1 = function1("ass")
result2 = function2(15)
result3 = function3(4)
result4 = function4(10)

print(f"Result from function1: {result1}")
print(f"Result from function2: {result2}")
print(f"Result from function3: {result3}")
print(f"Result from function4: {result4}")
```
